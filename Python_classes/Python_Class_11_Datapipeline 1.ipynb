```python
def getPubLinks(url):
    first_page_pub_links = []
    urlOpen =  urlopen(url)
    read_ton = soup(urlOpen.read())
    # Iterate over the HTML document 
    for link in read_ton.findAll("a"):
        if "href" in link.attrs:
            links_val = link.attrs["href"]
            documents_arr = links_val.split("/")
            if documents_arr[0] == "https:":
                if "publications" in documents_arr:
                    first_page_pub_links.append(links_val)
    return first_page_pub_links 
```

```python
# Download the Pdfs

def get_pdf_links(url):
    # Send a request to the URL and get the HTML content
    response = requests.get(url)
    response.raise_for_status()  # Raise an exception for HTTP errors

    # Parse the HTML content using Beautifulbs_soup
    bs_soup = soup(response.content, 'html.parser')

    # Find all links that end with .pdf
    pdf_links = [link.get('href') for link in bs_soup.find_all('a', href=True) if link['href'].lower().endswith('.pdf')]
    return pdf_links


def download_pdf(url, output_path):
    # Send a request to the URL and get the content
    response = requests.get(url)
    response.raise_for_status()  # Raise an exception for HTTP errors

    # Write the content to a file
    with open(output_path, 'wb') as file:
        file.write(response.content)
```

```python
for pdf_links in lnks:
        webpage_url= pdf_links
        pdf_links = get_pdf_links(webpage_url)
        print(pdf_links)

        # pdf_links
        # Ensure there's a directory to save the PDFs
        output_directory = r'/Users/admin/Documents/DataFolder_PDF'
        os.makedirs(output_directory, exist_ok=True)

        for pdf_link in pdf_links:
                # Handle relative URLs
                # full_pdf_link = f"{new_webpage}{pdf_links[0]}"
                full_pdf_link = urljoin(webpage_url, pdf_link)
                # print(full_pdf_link)
                
                # # Extract the PDF file name from the URL
                pdf_filename = os.path.basename(full_pdf_link) # Getting the file name 
                # print(pdf_filename)
                output_path = os.path.join(output_directory, pdf_filename)
                
                # # Download the PDF
                download_pdf(full_pdf_link, output_path)
                print(f"Downloaded {pdf_filename} to {output_path}")
```
