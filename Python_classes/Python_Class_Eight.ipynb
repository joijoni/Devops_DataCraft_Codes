```python
arr_firstname = []
arr_lastname = []

for i in arr:
    if "-" in i:
        split_data = i.split("-")
        arr_firstname.append(split_data[0])
        arr_lastname.append(split_data[-1])
        # pass
    else:
        split_data = re.split(splitters, i)
        arr_firstname.append(split_data[0])
        arr_lastname.append(split_data[-1])

dict_data = {"first name": arr_firstname, "last name":arr_lastname}
print(dict_data)
```

```python
#Put in a function

def handleNames(arr, splitters, first_value, last_value):
    arr_firstname = []
    arr_lastname = []

    for i in arr:
        if "-" in i:
            split_data = i.split("-")
            arr_firstname.append(split_data[first_value])
            arr_lastname.append(split_data[last_value])
            # pass
        else:
            split_data = re.split(splitters, i)
            arr_firstname.append(split_data[first_value])
            arr_lastname.append(split_data[last_value])

    dict_data = {"first name": arr_firstname, "last name":arr_lastname}
    return dict_data
```

```python
splitters = r'[#$/\–]'
dict_names = handleNames(arr, splitters, 0, -1)
dict_names
```
